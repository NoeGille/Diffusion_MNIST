```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return -3.8333*x + 46.056

def g(x):
    return 0.2568*x+24.56

# Plot the functions on the same graph
x = np.linspace(0, 18, 1000)
plt.plot(x, f(x), label='f(x) = -3.8333*x + 46.056', linestyle='--')
plt.plot(x, g(x), label='g(x) = 0.2568*x+24.56', linestyle='--')
plt.xlabel('Concentration')
plt.xticks(np.arange(0, 19, 2))
plt.ylabel('Force')

# Plot the intersection point and its value on the graph
z = np.abs(f(x) - g(x))
x0 = x[np.argmin(z)]
y0 = f(x0)
plt.plot(x0, y0, 'ro')
plt.text(x0, y0, '{}'.format(round(x0, ndigits=3)))
plt.title('Mesure de la concentration micellaire critique')
plt.grid(True)

# Add points 
p1 = [[]]
```
